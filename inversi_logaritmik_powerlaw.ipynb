```python
# Mengimport library yang digunakan
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load gambar dari local drive dengan cv2.imread
bombastic = cv2.imread("bombastic.jpg") 

# Mendapatkan resolusi dan type dari gambar
bombastic_height = bombastic.shape[0]     # Mengambil tinggi gambar dalam piksel, dan menyimpannya dalam variabel "bombastic_height" 
bombastic_width = bombastic.shape[1]      # Mengambil lebar gambar dalam piksel, dan menyimpannya dalam variabel "bombastic_width"
bombastic_channel = bombastic.shape[2]    # Mengambil jumlah channel warna dalam gambar, dan menyimpannya dalam variabel "bombastic_channel".

# PERCOBAAN PERTAMA : INVERSI
# Membuat variabel bombastic_inversi
bombastic_inversi = np.zeros(bombastic.shape, dtype=np.uint8)   # Membuat variabel 'bombastic_inversi' sebagai array numpy dengan tipe data uint8 yang berukuran sama dengan gambar input 'bombastic'.

# Membuat fungsi untuk inversi grayscale
def inversi_grayscale(nilai):              
    for y in range(0, bombastic_height):    # Mengiterasi setiap pixel pada gambar, kemudian menjumlahkan nilai intensitas dari ketiga komponen warna (merah, hijau, dan biru), kemudian dibagi dengan 3 untuk mendapatkan nilai grayscale.
        for x in range(0, bombastic_width):
            red = bombastic[y][x][0]
            green = bombastic[y][x][1]
            blue = bombastic[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            # Nilai grayscale tersebut akan di-inversi dengan cara mengurangi nilai maksimum (nilai 255) dengan nilai grayscale, kemudian hasil inversi tersebut disimpan ke dalam array 'bombastic_inversi'.
            gray = nilai - gray
            bombastic_inversi[y][x] = (gray, gray, gray)

# Membuat fungsi untuk inversi rgb
def inversi_rgb(nilai):
    for y in range(0, bombastic_height):    # Mengiterasi setiap pixel pada gambar, kemudian mengurangi nilai maksimum (nilai 255) dengan nilai masing-masing komponen warna (merah, hijau, dan biru).
        for x in range(0, bombastic_width):
            red = bombastic[y][x][0]
            red = nilai - red
            green = bombastic[y][x][1]
            green = nilai - green
            blue = bombastic[y][x][2]       # Hasil inversi tersebut disimpan ke dalam array 'bombastic_inversi'.
            blue = nilai - blue
            bombastic_inversi[y][x] = (red, green, blue)

# Menampilkan hasil inversi grayscale dan inversi RGB pada gambar 'bombastic', menggunakan fungsi imshow() dari library matplotlib.
inversi_grayscale(255)
plt.imshow(bombastic_inversi)
plt.title("Inversi Grayscale")
plt.show()

inversi_rgb(255)
plt.imshow(bombastic_inversi)
plt.title("Inversi RGB")
plt.show()
```

```python
# PERCOBAAN KEDUA : TRANSFORMASI LOGARITMIK

# Membuat variabel bombastic_log untuk menampung hasil
bombastic_log = np.zeros(bombastic.shape, dtype=np.uint8)

# Mendefinisikan fungsi untuk log
def log(c):
    for y in range(0, bombastic_height):
        for x in range(0, bombastic_width):
            red = bombastic[y][x][0]
            green = bombastic[y][x][1]
            blue = bombastic[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(c * np.log(gray + 1))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            bombastic_log[y][x] = (gray, gray, gray)

# Menampilkan hasil log
log(30)
plt.imshow(bombastic_log)
plt.title("Log")
plt.show()
```

```python
# PERCOBAAN KETIGA : INVERSI DAN LOG

# Membuat variabel bombastic_inlog untuk menampung hasil
bombastic_inlog = np.zeros(bombastic.shape, dtype=np.uint8)

# Mendefinisikan fungsi untuk inversi log
def inlog(c):
    for y in range(0, bombastic_height):
        for x in range(0, bombastic_width):
            red = bombastic[y][x][0]
            green = bombastic[y][x][1]
            blue = bombastic[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(c * np.log(255 - gray + 1))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            bombastic_inlog[y][x] = (gray, gray, gray)

# Menampilkan hasil inversi log
inlog(30)
plt.imshow(bombastic_inlog)
plt.title("Inversi & Log")
plt.show()
```

```python
# PERCOBAAN KEEMPAT : NTH POWER

# Membuat variabel bombastic_nthpower untuk menampung hasil
bombastic_nthpower = np.zeros(bombastic.shape, dtype=np.uint8)

# Mendefinisikan fungsi untuk nth power
def nthpower(c, y):
    thc = c / 100
    thy = y / 100
    for y in range(0, bombastic_height):
        for x in range(0, bombastic_width):
            red = bombastic[y][x][0]
            green = bombastic[y][x][1]
            blue = bombastic[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(thc * pow(gray, thy))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            bombastic_nthpower[y][x] = (gray, gray, gray)

# Menampilkan hasil
nthpower(50, 100)
plt.imshow(bombastic_nthpower)
plt.title("Nth Power")
plt.show()
```

```python
# PERCOBAAN KELIMA : NTH ROOT POWER

# Membuat variabel bombastic_nthrootpower
bombastic_nthrootpower = np.zeros(bombastic.shape, dtype=np.uint8)

# Membuat fungsi untuk nth root power
def nthrootpower(c, y):
    thc = c / 100
    thy = y / 100
    for y in range(0, bombastic_height):
        for x in range(0, bombastic_width):
            red = bombastic[y][x][0]
            green = bombastic[y][x][1]
            blue = bombastic[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(thc * pow(gray, 1./thy))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            bombastic_nthrootpower[y][x] = (gray, gray, gray)

# Menampilkan hasil
nthrootpower(50, 100)
plt.imshow(bombastic_nthrootpower)
plt.title("Nth Root Power")
plt.show()
```
