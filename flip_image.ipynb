```python
# Mengimport library yang digunakan
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load gambar dari local drive dengan cv2.imread
bombastic = cv2.imread("bombastic.jpg") 

# Mendapatkan resolusi dan type dari gambar
img_height = bombastic.shape[0]     # Mengambil tinggi gambar dalam piksel, dan menyimpannya dalam variabel "img_height" 
img_width = bombastic.shape[1]      # Mengambil lebar gambar dalam piksel, dan menyimpannya dalam variabel "img_width"
img_channel = bombastic.shape[2]    # Mengambil jumlah channel warna dalam gambar, dan menyimpannya dalam variabel "img_channel".
img_type = bombastic.dtype          # Mengambil tipe data gambar (misalnya, uint8, float32), dan menyimpannya dalam variabel "img_type".

# Membuat variabel dengan resolusi dan tipe yang sama seperti gambar
img_flip_horizontal = np.zeros(bombastic.shape, img_type) # Variabel gambar utk flip horizontal dengan resolusi dan type yg sama kemudian dikosongkan nilainya
img_flip_vertical = np.zeros(bombastic.shape, img_type)   # Variabel gambar utk flip vertical dengan resolusi dan type yg sama kemudian dikosongkan nilainya

# Membalik gambar secara horizontal
for y in range(0, img_height):              # Perulangan for variabel y (tinggi) dalam rentang 0 hingga "img_height"
    for x in range(0, img_width):           # Perulangan for variabel x (lebar) dalam rentang 0 hingga "img_width"
        for c in range(0, img_channel):     # Perulangan for variabel c (channel warna) dalam rentang 0 hingga "img_channel"
            img_flip_horizontal[y][x][c] = bombastic[y][img_width-1-x][c] # Mengalihkan indeks kolom sisi kiri ke kanan gambar dengan "img_width-1-x"

# Membalik gambar secara vertical
for y in range(0, img_height):              # Perulangan for variabel y (tinggi) dalam rentang 0 hingga "img_height"
    for x in range(0, img_width):           # Perulangan for variabel x (lebar) dalam rentang 0 hingga "img_width"
        for c in range(0, img_channel):     # Perulangan for variabel c (channel warna) dalam rentang 0 hingga "img_channel" 
            img_flip_vertical[y][x][c] = bombastic[img_height-1-y][x][c] # Mengalihkan indeks kolom sisi atas ke bawah gambar dengan "img_height-1-x"

# Menampilkan hasil balik gambar
fig, axes = plt.subplots(1, 3, figsize=(12, 12)) # Membuat figure subplot
ax = axes.ravel()                                # Mengubah array multi-dimensi yang merepresentasikan susunan subplot menjadi array satu dimensi       

# Plot gambar original dengan cv2.imshow
ax[0].imshow(cv2.cvtColor(bombastic, cv2.COLOR_BGR2RGB))    # Mengubah type gambar "bombastic" dari BGR ke RGB kemudian ditampilkan pada array axes pertama subplot 
ax[0].set_title("Original Image")                           # Memberikan judul gambar "Original Image"

# Plot gambar flip horizontal dengan cv2.imshow
ax[1].imshow(cv2.cvtColor(img_flip_horizontal, cv2.COLOR_BGR2RGB)) # Mengubah type gambar "img_flip_horizontal" dari BGR ke RGB kemudian ditampilkan pada array axes pertama subplot 
ax[1].set_title("Flip Horizontal Image")                           # Memberikan judul gambar "Flip Horizontal Image"

# Plot gambar flip horizontal dengan cv2.imshow
ax[2].imshow(cv2.cvtColor(img_flip_vertical, cv2.COLOR_BGR2RGB))   # Mengubah type gambar "img_flip_vertical" dari BGR ke RGB kemudian ditampilkan pada array axes pertama subplot 
ax[2].set_title("Flip Vertical Image")                             # Memberikan judul gambar "Flip Vertical Image"

fig.tight_layout()     # Menyesuaikan posisi elemen-elemen dalam plot atau gambar
plt.show()             # Menampilkan subplot gambar dengan matplotlib
```
